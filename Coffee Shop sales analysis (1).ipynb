```python
import pandas as pd
data = pd.read_excel('Coffee Shop Sales.xlsx')
data
```

```python
data.isna().sum()
```

```python
data['week day'] = data['transaction_date'].dt.strftime('%A')
data.head()
```

```python
data['time peroid'] = data['transaction_time'].apply(lambda x :  'morning' if  6<= x.hour < 12 else('afternoon' if 12 <= x.hour < 18 else 'evening'))
data.tail()
```

```python
data.groupby(data['transaction_date'].dt.strftime('%m'))['transaction_qty'].sum()
```

```python
data.groupby(data['time peroid'])['transaction_qty'].sum()
```

```python
data.groupby(data['store_location'])['transaction_qty'].count()
```

```python
data.product_category.unique()
```

```python
data.groupby('product_category')['transaction_qty'].sum()
```

```python
data.groupby('product_type')['transaction_qty'].count().sort_values(ascending = False)
```

```python
data.loc[data['time peroid'] == 'morning'].groupby('product_type')['transaction_qty'].count().sort_values()
```

```python
data.loc[data['time peroid'] == 'afternoon'].groupby('product_type')['transaction_qty'].count().sort_values()
```

```python
data.loc[data['time peroid'] == 'evening'].groupby('product_type')['transaction_qty'].count().sort_values()
```

```python
data.groupby('week day')['transaction_qty'].sum().sort_values(ascending = False)
```
